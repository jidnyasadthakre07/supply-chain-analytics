#### Phase 4b — KPI Deep Dive
Interactive exploration of all KPI outputs.

```python
import sys, pandas as pd, matplotlib.pyplot as plt
sys.path.append('../src')
from kpi import kpi_executive_summary, kpi_by_region, kpi_by_product, kpi_monthly_trend
```

```python
summary = kpi_executive_summary()
for k, v in summary.items():
    print(f'{k:<28}: {v}')
```

```python
region_df = kpi_by_region()
region_df.plot(kind='bar', x='region_name', y='total_revenue',
               color='#534AB7', figsize=(8,4), legend=False)
plt.title('Revenue by Region')
plt.tight_layout()
plt.show()
```

```python
product_df = kpi_by_product()
product_df.plot(kind='barh', x='product_name', y='total_revenue',
                color='#1D9E75', figsize=(8,5), legend=False)
plt.title('Revenue by Product')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
trend_df = kpi_monthly_trend()
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(range(len(trend_df)), trend_df['total_revenue'],
        marker='o', color='#534AB7', linewidth=2)
ax.set_xticks(range(len(trend_df)))
ax.set_xticklabels(trend_df['month_name'].tolist(), rotation=45)
ax.set_title('Monthly Revenue Trend')
plt.tight_layout()
plt.show()
```
